## Prediction using saved model

Specifying corrector with model name (default name:"model"):

```python
corrector = AECorrector(model_name="my_model1")
```

Training and saving the model:

```python
corrected = corrector.correct(data)
```

Only predicting, using model "my_model1" saved on disk:

```python
corrected = corrector.correct(data,only_predict=True)
```

```python
ev = Evaluation(data, corrected)
```

```python
ev.plot_correlations_of_samples()
```
